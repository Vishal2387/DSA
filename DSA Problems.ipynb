1. Given an array A of size N of integers. Your task is to find the minimum and maximum elements in the array

```python
def max(a,n):
    max = a[0]
    for i in range(n):
        if a[i] > max:
            max = a[i]
    return max

def min(a,n):
    min = a[0]
    for i in range(n):
        if a[i] < min:
            min = a[i]
    return min

a = [3, 2, 1, 56, 10000, 167]
n = len(a)
def getMinMax( a, n):
    return min(a,n), max(a,n)
getMinMax(a,n)
```

2. Given an array of distinct elements. Find the third largest element in it.

```python
def thirdLargest(a, n):
    fmax = a[0]
    smax = 0
    tmax = -1
    for i in range (len(a)):
        if a[i] > fmax: 
            tmax = smax
            smax = fmax
            fmax = a[i]  
        elif a[i] < fmax and a[i] > smax:
            tmax = smax
            smax = a[i]
        elif a[i] < fmax and a[i] < smax and a[i] > tmax :
            tmax = a[i]
    return tmax
a= [884, 337, 689, 587, 748, 308, 451, 785, 682, 600, 733, 537, 403, 188, 162, 562, 274, 513, 383, 617, 662, 508, 111, 57, 160]
n = len(a)
thirdLargest(a, n)
```

3. Given an integer array and another integer element. The task is to find if the given element is present in array or not.

```python
def search(arr, N, X):
    for i in range (N):
        if arr[i] == X:
            op = i
            break
        else:
            op = -1
    return op
arr = [1,2,3,4,5]
N = 5
X = 5
search(arr, N, X)
```

4. Given an array of size N-1 such that it only contains distinct integers in the range of 1 to N. Find the missing element.

```python
def missingNumber(array,n):
    sum_n = n * (n+1)//2
    sum_arr = 0
    for i in range(len(array)):
        sum_arr = sum_arr + array[i]
    op = sum_n - sum_arr
    return op
arr = [1,3,4,5,6]
n = 6
missingNumber(arr,n)
```

5. Given an array arr, rotate the array of k

```python
def rotate(arr, s, e):
    while s<e:
        t = arr[s]
        arr[s] = arr[e]
        arr[e] = t
        s = s+1
        e = e-1
arr = [1,2,3,4,5]
k = 3
rotate(arr, 0, len(arr)-1)
rotate(arr, 0, k-1)
rotate(arr, k, len(arr)-1)
print(arr)
```

## Two, Three four Pointers

6. Given an array arr[] of n integers. Check whether it contains a triplet that sums up to zero. Return 1, if there is at least one triplet following the condition else return false. (use 3 pointers technique i, j, k)

```python
def findTriplets(arr, n):
    arr.sort()
    for i in range (n):
        j = i +1
        k = n -1
        while j<k:
            sum = arr[i]+arr[j]+arr[k]
            if sum == 0:
                return 1
            elif sum < 0:
                j = j+1
            else:
                k = k-1
arr = [0, -1, 2, -3, 1]
n = 5
findTriplets(arr,n)
```

7. Given an array Arr consisting of N distinct integers. The task is to count all the triplets such that sum of two elements equals the third element.

```python
def countTriplet(arr, n):
    arr.sort()
    count = 0
    for k in reversed(range(n)):
	    i = 0
	    j = k-1
	    while(i<j):
	        if arr[k] == arr[i]+arr[j]:
	            count = count +1
	            i = i + 1
	            j = j - 1
	        elif arr[k] > arr[i]+arr[j]:
	            i = i + 1
	        else:
	            j = j - 1
    return count
arr = [7, 3, 2 ,5, 4, 1]
n = 6
countTriplet(arr,n)
```

8. The majority element is the element that appears more than ⌊n / 2⌋ times. You may assume that the majority element always exists in the array.

```python
def majorityElement(nums):
    n = len (nums) 
    count = 0       
    for i in range(n):
        if count == 0:
            count = 1
            ele = nums[i]
        elif ele == nums[i]:
            count += 1
        else:
            count -= 1
    count1 = 0
    for i in range(n):
        if ele == nums[i]:
            count1 += 1
        if count1 > n/2:
            return ele
nums = [2,2,1,1,1,2,2]
majorityElement(nums)
```

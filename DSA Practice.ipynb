```python
arr = [1,2,3,4,5]
temp = arr[0]
for i in range(len(arr)-1):
    arr[i] = arr[i+1]
arr[len(arr)-1] = temp
print(arr)
```

```python
arr = [1,2,3,4,5]
d = 3
temp = arr[0:d]
for i in range(len(arr)-d):
    arr[i] = arr[i+d]
arr[len(arr)-d:] = temp
print(arr)
```

```python
# Move all zeros to the end of the array
def move_zeros(arr):
    j = 0
    for i in range(len(arr)):
        if arr[i] != 0:
            arr[j] = arr[i]
            j += 1
    for k in range(j, len(arr)):
        arr[k] = 0
    return arr

arr = [1,0,2,3,2,0,0,4,5,1]
print(move_zeros(arr))
```

```python
def findUnion1(a, b):
    i, j = 0, 0
    a,b = sorted(a), sorted(b)
    op = []
    while i < len(a) and j < len(b):
        if a[i] <= b[j]:
            if len(op) == 0 or op[-1] != a[i]:
                op.append(a[i])
            i = i + 1    
        else:
            if len(op) == 0 or op[-1] != b[j]:
                op.append(b[j])
            j = j + 1    
    while i < len(a):
        if len(op) == 0 or op[-1] != a[i]:
            op.append(a[i])
        i = i + 1    
    while j < len(b):
        if len(op) == 0 or op[-1] != b[j]:
            op.append(b[j])
        j = j + 1    
    return sorted(op)
a = [10,2,6,3]
b = [2,10,10,8]
print(findUnion1(a, b))
```

```python
# Intersection of two unsorted arrays
def findIntersection(a, b):
    a = set(a)
    b = set(b)
    op = []
    op = a & b
    return len(op)
a = [10,2,6,3]  
b = [2,10,10,8]
print(findIntersection(a, b))
```

```python
# Intersection of two unsorted arrays if the arrays have distinct elements
def findIntersection(a, b):
    op = []
    a = set(a)
    for i in range(len(b)):
        if b[i] in a:
            op.append(b[i])
    return op
a = [10,2,6,3]  
b = [2,10,8]
print(findIntersection(a, b))
```
